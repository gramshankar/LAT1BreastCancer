```
import pandas as pd
from scipy.stats import mannwhitneyu
```

```
data = pd.read_csv("Log2FC Mann Whitney Data - Postmeno.csv", usecols = [i for i in range(1,22,2)])
Ki67 = data["Post Ki-67"].drop([1,2,11,15,21,23,25,26]).dropna()
Basophil = data["Post Basophil"].drop([1,10,19]).dropna()
Eosinophil = data["Post Eosinophil"].drop([1,10,19]).dropna()
Neutrophil = data["Post Neutrophil"].drop([1,10,19]).dropna()
Monocyte = data["Post Monocyte"].drop([1,10,19]).dropna()
WBC = data["Post WBC"].drop([1,10,19]).dropna()
Lymphocyte = data["Post Lymphocyte"].drop([1,10,19]).dropna()
```

```
mannwhitneyu(Ki67,Basophil)
```

```
mannwhitneyu(Ki67,Eosinophil)
```

```
mannwhitneyu(Ki67,Neutrophil)
```

```
mannwhitneyu(Ki67,Monocyte)
```

```
mannwhitneyu(Ki67,WBC)
```

```
mannwhitneyu(Ki67,Lymphocyte)
```

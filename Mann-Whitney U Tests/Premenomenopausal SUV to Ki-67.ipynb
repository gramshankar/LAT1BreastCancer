```
import pandas as pd
from scipy.stats import mannwhitneyu
```

```
data = pd.read_csv("Log2FC Mann Whitney Data - Premeno.csv", usecols = [i for i in range(1,22,2)])
SUVMean = data["Pre SUV Mean"].drop([17,23])
SUVPeak = data["Pre SUV Peak"].drop([17,23])
SUVMax = data["Pre SUV Max"].drop([17,23])
Ki67 = data["Pre Ki-67"].dropna()
```

```
mannwhitneyu(SUVMean,Ki67)
```

```
mannwhitneyu(SUVPeak,Ki67)
```

```
mannwhitneyu(SUVMax,Ki67)
```

```
import pandas as pd
from scipy.stats import mannwhitneyu
```

```
data = pd.read_csv("Log2FC Mann Whitney Data - Premeno.csv", usecols = [i for i in range(1,22,2)])
BMI = data["Pre BMI"].drop([1,8,9,10,12,16,17,18,19,23,25])
Basophil = data["Pre Basophil"].dropna()
Eosinophil = data["Pre Eosinophil"].dropna()
Neutrophil = data["Pre Neutrophil"].dropna()
Monocyte = data["Pre Monocyte"].dropna()
WBC = data["Pre WBC"].dropna()
Lymphocyte = data["Pre Lymphocyte"].dropna()
```

```
mannwhitneyu(BMI,Basophil)
```

```
mannwhitneyu(BMI,Eosinophil)
```

```
mannwhitneyu(BMI,Neutrophil)
```

```
mannwhitneyu(BMI,Monocyte)
```

```
mannwhitneyu(BMI,WBC)
```

```
mannwhitneyu(BMI,Lymphocyte)
```

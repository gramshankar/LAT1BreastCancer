```
import pandas as pd
from scipy.stats import mannwhitneyu
```

```
data = pd.read_csv("Log2FC Mann Whitney Data - Postmeno.csv", usecols = [i for i in range(1,22,2)])
SUVMean = data["Post SUV Mean"].drop([2,3,12,17,18,23,27,29,30])
SUVPeak = data["Post SUV Peak"].drop([2,3,12,17,18,23,27,29,30])
SUVMax = data["Post SUV Max"].drop([2,3,12,17,18,23,27,29,30])
Basophil = data["Post Basophil"].dropna()
Eosinophil = data["Post Eosinophil"].dropna()
Neutrophil = data["Post Neutrophil"].dropna()
Monocyte = data["Post Monocyte"].dropna()
WBC = data["Post WBC"].dropna()
Lymphocyte = data["Post Lymphocyte"].dropna()
```

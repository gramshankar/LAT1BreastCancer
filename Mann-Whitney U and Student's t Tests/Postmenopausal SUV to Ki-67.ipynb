```
import pandas as pd
from scipy.stats import mannwhitneyu
```

```
data = pd.read_csv("Log2FC Mann Whitney Data - Postmeno.csv", usecols = [i for i in range(1,22,2)])
SUVMean = data["Post SUV Mean"].drop([1,13,18,23])
SUVPeak = data["Post SUV Peak"].drop([1,13,18,23])
SUVMax = data["Post SUV Max"].drop([1,13,18,23])
Ki67 = data["Post Ki-67"].dropna()
```

```
mannwhitneyu(SUVMean,Ki67)
```

```
mannwhitneyu(SUVPeak,Ki67)
```

```
mannwhitneyu(SUVMax,Ki67)
```

```
import pandas as pd
from scipy.stats import pearsonr
```

```
data = pd.read_csv('All SUV Values - Included Postmenopausal SUV Max.csv')
data = data.drop([3,18,19,24,30,31])
```

```
data = data.set_index("ID")
```

```
he = pd.read_csv("HE.csv")
Immune = he.loc[[4,7,18,21,23,25,29,32,37,40,41,42,47,48,49,51,59,61,63,66,69,77,79,81,82,89], :].set_index("cn")
```

```
basophil = Immune["hee124"]
eosinophil = Immune["hee115"]
neutrophil = Immune["hee133"]
monocyte = Immune["hee106"]
WBC = Immune["hee25"]
lymphocyte = Immune["hee97"]
```

```
data["basophil"] = basophil
data["eosinophil"] = eosinophil
data["neutrophil"] = neutrophil
data["monocyte"] = monocyte
data["WBC"] = WBC
data["lymphocyte"] = lymphocyte
```

```
scan1 = data["SUV Max Scan 1"]
scan2 = data["SUV Max Scan 2"]
scan3 = data["SUV Max Scan 3"]
scan4 = data["SUV Max Scan 4"]
scan5 = data["SUV Max Scan 5"]
```

```
Scan1 = []
for scan in scan1:
    scana = str(scan)
    split1 = scana.split()
    val = split1.pop(0)
    Scan1.append(float(val))
print(Scan1)
```

```
Scan2 = []
for scan in scan2:
    scana = str(scan)
    split2 = scana.split()
    val = split2.pop(0)
    Scan2.append(float(val))
print(Scan2)
```

```
Scan3 = []
for scan in scan3:
    scana = str(scan)
    split3 = scana.split()
    val = split3.pop(0)
    Scan3.append(float(val))
print(Scan3)
```

```
Scan4 = []
for scan in scan4:
    scana = str(scan)
    split4 = scana.split()
    val = split4.pop(0)
    Scan4.append(float(val))
print(Scan4)
```

```
Scan5 = []
for scan in scan5:
    scana = str(scan)
    split5 = scana.split()
    val = split5.pop(0)
    Scan5.append(float(val))
print(Scan5)
```

```
scan12 = [a + b for a,b in zip(Scan1,Scan2)]
scan34 = [a + b for a,b in zip(Scan3,Scan4)]
scan125 = [a + b for a,b in zip(scan12,Scan5)]
total = [a + b for a,b in zip(scan125,scan34)]
```

```
SUVMaxAverage = [a/5 for a in total]
```

```
basophilA = basophil.tolist()
eosinophilA = eosinophil.tolist()
neutrophilA = neutrophil.tolist()
monocyteA = monocyte.tolist()
WBCA = WBC.tolist()
lymphocyteA = lymphocyte.tolist()
```

```
pearsonr(SUVMaxAverage,basophilA)
```

```
pearsonr(SUVMaxAverage,eosinophilA)
```

```
pearsonr(SUVMaxAverage,neutrophilA)
```

```
pearsonr(SUVMaxAverage,monocyteA)
```

```
pearsonr(SUVMaxAverage,WBCA)
```

```
pearsonr(SUVMaxAverage,lymphocyteA)
```

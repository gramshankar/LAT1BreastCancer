```
import pandas as pd
from scipy.stats import pearsonr
```

```
data = pd.read_csv('All SUV Values - Included Postmenopausal SUV Max.csv', skiprows=[2,14,19,26])
```

```
data = data.set_index("ID")
```

```
ph = pd.read_csv("PH.csv")
values = ph.loc[[4,16,17,20,22,24,28,31,36,39,40,41,46,47,49,54,57,59,61,64,66,67,77,79,80,84,85,86], :].set_index("cn")
display(ph.head(90))
display(values.head(90))
```

```
ki67 = values["phe61"]
ki67
```

```
data["ki67"] = ki67
```

```
data.head()
```

```
scan1 = data["SUV Max Scan 1"]
scan2 = data["SUV Max Scan 2"]
scan3 = data["SUV Max Scan 3"]
scan4 = data["SUV Max Scan 4"]
scan5 = data["SUV Max Scan 5"]
```

```
Scan1 = []
for scan in scan1:
    scana = str(scan)
    split1 = scana.split()
    val = split1.pop(0)
    Scan1.append(float(val))
print(Scan1)
```

```
Scan2 = []
for scan in scan2:
    scana = str(scan)
    split2 = scana.split()
    val = split2.pop(0)
    Scan2.append(float(val))
print(Scan2)
```

```
Scan3 = []
for scan in scan3:
    scana = str(scan)
    split3 = scana.split()
    val = split3.pop(0)
    Scan3.append(float(val))
print(Scan3)
```

```
Scan4 = []
for scan in scan4:
    scana = str(scan)
    split4 = scana.split()
    val = split4.pop(0)
    Scan4.append(float(val))
print(Scan4)
```

```
Scan5 = []
for scan in scan5:
    scana = str(scan)
    split5 = scana.split()
    val = split5.pop(0)
    Scan5.append(float(val))
print(Scan5)
```

```
scan12 = [a + b for a,b in zip(Scan1,Scan2)]
scan34 = [a + b for a,b in zip(Scan3,Scan4)]
scan125 = [a + b for a,b in zip(scan12,Scan5)]
total = [a + b for a,b in zip(scan125,scan34)]
```

```
SUVMaxAverage = [a/5 for a in total]
```

```
Ki67 = ki67.tolist()
```

```
pearsonr(SUVMaxAverage,Ki67)
```

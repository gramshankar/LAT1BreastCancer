```
import pandas as pd
from scipy.stats import pearsonr
```

```
data = pd.read_csv('All SUV Values - Included Postmenopausal SUV Mean.csv')
data = data.drop([3,18,19,24,30,31])
```

```
data = data.set_index("ID")
```

```
he = pd.read_csv("HE.csv")
Immune = he.loc[[4,7,18,21,23,25,29,32,37,40,41,42,47,48,49,51,59,61,63,66,69,77,79,81,82,89], :].set_index("cn")
```

```
basophil = Immune["hee124"]
eosinophil = Immune["hee115"]
neutrophil = Immune["hee133"]
monocyte = Immune["hee106"]
WBC = Immune["hee25"]
lymphocyte = Immune["hee97"]
```

```
data["basophil"] = basophil
data["eosinophil"] = eosinophil
data["neutrophil"] = neutrophil
data["monocyte"] = monocyte
data["WBC"] = WBC
data["lymphocyte"] = lymphocyte
```

```
scan1 = data["SUV Mean Scan 1 "]
scan2 = data["SUV Mean Scan 2"]
scan3 = data["SUV Mean Scan 3"]
scan4 = data["SUV Mean Scan 4"]
scan5 = data["SUV Mean Scan 5"]
```

```
upperScan1 = []
lowerScan1 = []
for scan in scan1:
    scana = str(scan)
    split1 = scana.split()
    upperScan1.append(float(split1[0]) + float(split1[2]))
    lowerScan1.append(float(split1[0]) - float(split1[2]))
```

```
upperScan2 = []
lowerScan2 = []
for scan in scan2:
    scana2 = str(scan)
    split2 = scana.split()
    upperScan2.append(float(split2[0]) + float(split2[2]))
    lowerScan2.append(float(split2[0]) - float(split2[2]))
```

```
upperScan3 = []
lowerScan3 = []
for scan in scan3:
    scana3 = str(scan)
    split3 = scan.split()
    upperScan3.append(float(split3[0]) + float(split3[2]))
    lowerScan3.append(float(split3[0]) - float(split3[2]))
```

```
upperScan4 = []
lowerScan4 = []
for scan in scan4:
    scana4 = str(scan)
    split4 = scan.split()
    upperScan4.append(float(split4[0]) + float(split4[2]))
    lowerScan4.append(float(split4[0]) - float(split4[2]))
```

```
upperScan5 = []
lowerScan5 = []
for scan in scan5:
    scana5 = str(scan)
    split5 = scan.split()
    upperScan5.append(float(split5[0]) + float(split5[2]))
    lowerScan5.append(float(split5[0]) - float(split5[2]))
```

```
combinedScan1 = [a + b for a,b in zip(upperScan1,lowerScan1)]
combinedScan2 = [a + b for a,b in zip(upperScan2,lowerScan2)]
combinedScan3 = [a + b for a,b in zip(upperScan3,lowerScan3)]
combinedScan4 = [a + b for a,b in zip(upperScan4,lowerScan4)]
combinedScan5 = [a + b for a,b in zip(upperScan5,lowerScan5)]
```

```
scan12 = [a + b for a,b in zip(combinedScan1,combinedScan2)]
scan34 = [a + b for a,b in zip(combinedScan3,combinedScan4)]
scan125 = [a + b for a,b in zip(scan12,combinedScan5)]
total = [a + b for a,b in zip(scan125,scan34)]
```

```
SUVMeanAverage = [a/10 for a in total]
```

```
basophilA = basophil.tolist()
eosinophilA = eosinophil.tolist()
neutrophilA = neutrophil.tolist()
monocyteA = monocyte.tolist()
WBCA = WBC.tolist()
lymphocyteA = lymphocyte.tolist()
```

```
pearsonr(SUVMeanAverage,basophilA)
```

```
pearsonr(SUVMeanAverage,eosinophilA)
```

```
pearsonr(SUVMeanAverage,neutrophilA)
```

```
pearsonr(SUVMeanAverage,monocyteA)
```

```
pearsonr(SUVMeanAverage,WBCA)
```

```
pearsonr(SUVMeanAverage,lymphocyteA)
```

```
import pandas as pd
from scipy.stats import pearsonr
```

```
data = pd.read_csv('All SUV Values - Included Postmenopausal SUV Peak.csv')
```

```
data = data.set_index("ID")
```

```
ft = pd.read_csv("FT.csv")
FirstVisit = ft.loc[[11,20,48,49,58,64,70,82,91,106,115,118,121,134,137,140,146,161,167,170,176,182,191,197,200,224,230,236,239,251,257,260], :].set_index("cn")
display(ft.head(265))
display(FirstVisit.head(90))
```

```
weight = FirstVisit["fte30"]
print(weight)
```

```
height = FirstVisit["fte32"]
print(height)
```

```
bmi = weight/(height/100)**2
```

```
bmi
```

```
data["bmi"] = bmi
```

```
data.head()
```

```
scan1 = data["SUV Peak Scan 1"]
scan2 = data["SUV Peak Scan 2"]
scan3 = data["SUV Peak Scan 3"]
scan4 = data["SUV Peak Scan 4"]
scan5 = data["SUV Peak Scan 5"]
```

```
Scan1 = []
for scan in scan1:
    scana = str(scan)
    split1 = scana.split()
    val = split1.pop(0)
    Scan1.append(float(val))
print(Scan1)
```

```
Scan2 = []
for scan in scan2:
    scana = str(scan)
    split2 = scana.split()
    val = split2.pop(0)
    Scan2.append(float(val))
print(Scan2)
```

```
Scan3 = []
for scan in scan3:
    scana = str(scan)
    split3 = scana.split()
    val = split3.pop(0)
    Scan3.append(float(val))
print(Scan3)
```

```
Scan4 = []
for scan in scan4:
    scana = str(scan)
    split4 = scana.split()
    val = split4.pop(0)
    Scan4.append(float(val))
print(Scan4)
```

```
Scan5 = []
for scan in scan5:
    scana = str(scan)
    split5 = scana.split()
    val = split5.pop(0)
    Scan5.append(float(val))
print(Scan5)
```

```
scan12 = [a + b for a,b in zip(Scan1,Scan2)]
scan34 = [a + b for a,b in zip(Scan3,Scan4)]
scan125 = [a + b for a,b in zip(scan12,Scan5)]
total = [a + b for a,b in zip(scan125,scan34)]
```

```
SUVPeakAverage = [a/5 for a in total]
```

```
BMI = bmi.tolist()
```

```
pearsonr(SUVPeakAverage,BMI)
```

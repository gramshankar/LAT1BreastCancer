```
import pandas as pd
from scipy.stats import pearsonr
```

```
ft = pd.read_csv("FT.csv")
FirstVisit = ft.loc[[0,3,5,14,17,24,27,52,61,73,76,79,85,94,112,152,173,209,215,218,221,227,242,248], :].set_index("cn")
display(FirstVisit.head(90))
```

```
ph = pd.read_csv("PH.csv")
values = ph.loc[[0,1,2,5,6,8,9,18,21,25,26,27,29,32,38,51,58,70,72,73,74,76,81,83], :].set_index("cn")
display(values.head(90))
```

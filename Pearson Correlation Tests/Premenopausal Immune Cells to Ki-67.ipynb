```
import pandas as pd
from scipy.stats import pearsonr
```

```
data = pd.read_csv('All SUV Values - Included Premenopausal SUV Max.csv', skiprows=[18,24])
```

```
data = data.set_index("ID")
```

```
ph = pd.read_csv("PH.csv")
values = ph.loc[[0,2,5,6,8,9,18,26,27,32,38,51,70,73,74,76,81], :].set_index("cn")
```

```
he = pd.read_csv("HE.csv")
Immune = he.loc[[0,2,5,6,9,10,19,27,28,33,39,53,72,75,76,78,83], :].set_index("cn")
```

```
basophil = Immune["hee124"]
eosinophil = Immune["hee115"]
neutrophil = Immune["hee133"]
monocyte = Immune["hee106"]
WBC = Immune["hee25"]
lymphocyte = Immune["hee97"]
data["basophil"] = basophil
data["eosinophil"] = eosinophil
data["neutrophil"] = neutrophil
data["monocyte"] = monocyte
data["WBC"] = WBC
data["lymphocyte"] = lymphocyte
```

```
ki67 = values["phe61"]
ki67
```

```
data["ki67"] = ki67
```

```
Ki67 = ki67.tolist()
```

```
basophilA = basophil.tolist()
eosinophilA = eosinophil.tolist()
neutrophilA = neutrophil.tolist()
monocyteA = monocyte.tolist()
WBCA = WBC.tolist()
lymphocyteA = lymphocyte.tolist()
```

```
pearsonr(Ki67,basophilA)
```

```
pearsonr(Ki67,eosinophilA)
```

```
pearsonr(Ki67,neutrophilA)
```

```
pearsonr(Ki67,monocyteA)
```

```
pearsonr(Ki67,WBCA)
```

```
pearsonr(Ki67,lymphocyteA)
```

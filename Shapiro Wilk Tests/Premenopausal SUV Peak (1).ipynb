```
import pandas as pd
from scipy.stats import shapiro
```

```
data = pd.read_csv('All SUV Values - Included Premenopausal SUV Peak.csv')
```

```
data = data.set_index("ID")
```

```
data.head()
```

```
scan1 = data["SUV Peak Scan 1"]
scan2 = data["SUV Peak Scan 2"]
scan3 = data["SUV Peak Scan 3"]
scan4 = data["SUV Peak Scan 4"]
scan5 = data["SUV Peak Scan 5"]
```

```
Scan1 = []
for scan in scan1:
    scana = str(scan)
    split1 = scana.split()
    val = split1.pop(0)
    Scan1.append(float(val))
print(Scan1)
```

```
Scan2 = []
for scan in scan2:
    scana = str(scan)
    split2 = scana.split()
    val = split2.pop(0)
    Scan2.append(float(val))
print(Scan2)
```

```
Scan3 = []
for scan in scan3:
    scana = str(scan)
    split3 = scana.split()
    val = split3.pop(0)
    Scan3.append(float(val))
print(Scan3)
```

```
Scan4 = []
for scan in scan4:
    scana = str(scan)
    split4 = scana.split()
    val = split4.pop(0)
    Scan4.append(float(val))
print(Scan4)
```

```
Scan5 = []
for scan in scan5:
    scana = str(scan)
    split5 = scana.split()
    val = split5.pop(0)
    Scan5.append(float(val))
print(Scan5)
```

```
scan12 = [a + b for a,b in zip(Scan1,Scan2)]
scan34 = [a + b for a,b in zip(Scan3,Scan4)]
scan125 = [a + b for a,b in zip(scan12,Scan5)]
total = [a + b for a,b in zip(scan125,scan34)]
```

```
SUVPeakAverage = [a/5 for a in total]
```

```
shapiro(SUVPeakAverage)
```
